### 1st technique for missing categorical features



```python
data = dataset.copy()
def replace_cat_features(data, categorical_features):
    
    data[categorical_features] = data[categorical_features].fillna('Missing')
    return data

```

```python
replace_cat_features(data, categorical_features)
```

### 2nd technique for missing categorical featuresm

```python
dataset['LotFrontage'].mode()
```

```python
data = dataset.copy()
def replace_cat_features(dataset, categorical_features):
    
    data[categorical_features] = data[categorical_features].fillna(dataset['LotFrontage'].mode())
    return data
```

```python
replace_cat_features(data, categorical_features)
```

# Numerical values

```python
numerical_features_nan = [feature for feature in dataset.columns if dataset[feature].dtypes != 'O' 
                            and dataset[feature].isnull().sum()>=1]
```

```python
numerical_features_nan
```

```python
dataset['LotFrontage'].median()
```

```python
for feature in numerical_features_nan:
    # since there are outliers, replace with median
    
    median_values = dataset[feature].median()
    
    # create a new feature to capture NaN values
    dataset[feature+'nan'] = np.where(dataset[feature].isnull(), 1,0)
    dataset[feature].fillna(median_values,inplace=True)
```
